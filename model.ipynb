```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model as lm
import pickle
from sklearn.externals import joblib
```

```python
df=pd.read_csv("insuranc.csv")
df.head()
```

```python
df.describe()
```

```python
value=np.random.rand(len(df))<.8
train=df[value]
test=df[~value]
```

```python
print(train)
```

```python
print(test)
```

```python
regr=lm.LinearRegression()
train_x=np.asanyarray(train[["age","bmi"]])
train_y=np.asanyarray(train[["charges"]])
regr.fit(train_x,train_y)
```

```python
print(regr.coef_)
```

```python
print(regr.intercept_)
```

```python
#Saving the model
joblib.dump(regr,"model.pkl")
```

```python
test_x=np.asanyarray(test[["age","bmi"]])
test_y=np.asanyarray(test[["charges"]])
prediction=regr.predict(test_x)
print(prediction)
```
